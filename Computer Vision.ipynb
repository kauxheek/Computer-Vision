```python
import cv2 as cv
import numpy as np
from playsound import playsound
```

```python

ip = "https://xx.xx.xx.43:8080"
video = cv.VideoCapture(ip+'/video')
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE,(3,3))
fgbg = cv.createBackgroundSubtractorMOG2()

while True:
    ret , frame = video.read()
    fgmask = fgbg.apply(np.array(frame))
    fgmask = cv.morphologyEx(fgmask,cv.MORPH_OPEN,kernel)
    #gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    cv.imshow('frame',fgmask)
    if cv.waitKey(1) == ord('q'):
        break
        
video.release()
cv.destroyAllWindows()
```

```python
body_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
ip = "https://xx.xx.xx.120:8080"
video = cv.VideoCapture(ip+'/video')
is_detected = False
while True:
    ret,frame = video.read()
    frame = np.array(frame)
    print(frame)
    gray = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    body  = body_cascade.detectMultiScale(gray,1.5,5)
    
        
    if len(body)>0:
        
        for (x,y,w,h) in body:
    
            frame = cv.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        is_detected = True
        if is_detected == True:
            playsound('speech.wav')
            is_detected == False
        
        
        
    cv.imshow('frame',frame)
    if cv.waitKey(1) == ord('q'):
        break
    
    

video.release()
cv.destroyAllWindows()
    
```

```python
car_cascade = cv.CascadeClassifier('cars.xml')
car  = cv.VideoCapture('carss.mp4')
if car.isOpened():
    midpoint =  car.get(3)/2
    height = car.get(4)



while True:
    ret,frame = car.read()
    frame = np.array(frame)
    gray = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    car_body = car_cascade.detectMultiScale(gray,1.1,6)
    for (x,y,w,h) in  car_body:
        frame = cv.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
        mid_x = (2*x+w)/2
        mid_y = y+h
        #distance  = np.round(np.sqrt((mid_x-midpoint)*(mid_x-midpoint)+(mid_y-height)*(mid_y-height)))
        distance = height-mid_y
        angle = str(np.round(np.degrees(np.arctan((0.5*((midpoint-mid_x)/(height-mid_y)))))))
        text = str(distance)
        cv.putText(frame,text,(x+w,y+h),cv.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
        cv.putText(frame,angle,(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
        
        print(angle)
    cv.imshow('test',frame)
    if cv.waitKey(1) == ord('q'):
            break
    
    

car.release()
cv.destroyAllWindows()

```

```python
video = cv.VideoCapture('carss.mp4')
while True:
    ret, frame = video.read()
    hsv = cv.cvtColor(frame,cv.COLOR_BGR2HSV)
    
    lower_blue = np.array([0,0,0])
    upper_blue = np.array([0,0,50])
    
    mask = cv.inRange(hsv,lower_blue,upper_blue)
    res = cv.bitwise_and(frame,frame,mask=mask)
    cv.imshow('frame',frame)
    cv.imshow('mask',mask)
    cv.imshow('res',res)
    k = cv.waitKey(5) & 0xFF
    if k == 27:
        break
cv.destroyAllWindows()
    
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
#img  = cv2.VideoCapture('carss.mp4')
ip = "https://10.90.230.23:8080"
img = cv2.VideoCapture(ip+'/video')
while True:
    ret,frame = img.read()
    mask = np.zeros(frame.shape[:2],np.uint8)
    bgdModel = np.zeros((1,65),np.float64)
    fgdModel = np.zeros((1,65),np.float64)
    rect = (50,50,450,290)
    cv2.grabCut(frame,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
    mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
    frame1 = frame*mask2[:,:,np.newaxis]
    cv2.imshow('ddd',frame1)
    cv2.imshow('dd',frame)
    if cv2.waitKey(1) == ord('q'):
        break
img.release()
cv2.destroyAllWindows()       
```
